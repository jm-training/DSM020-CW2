## Judging a book by its synopsis - a proposal for an automated classification system

In the world of book retailing, the task of categorising new books correctly is vital to maximising both audience reach and sales.  Poor classification could misrepresent the book’s content in its metadata and not only impact its in-store shelve location but also its chances of being surfaced correctly in online searches and recommendations. 

In the UK it’s generally left to the author or the publisher to categorise a book manually by following the guidelines of the Book Industry Communication (BIC) categorisation system (BIC and Nielsen Book Services Ltd, 2010). Such a task needs to consider how to treat a book’s subject matter amongst many of BIC’s 19 categories and dozens of various sub-categories.  

By considering NLP techniques to manipulate text data and machine learning algorithms to fit models to the transformed data it may be possible to create an accurate classifier tool which could remove this manual process. A quick literature search using Summon shows that there are many such use cases.  As I found the NLP techniques introduced in Topic 1 ‘Data structures’ very interesting I propose to undertake my own use case, a proof of concept, using book metadata captured from a High street retailer.  

This notebook details the necessary activity to see if such a proposal is valid.

- A web crawler was created to extract the data. Bestselling books were selected in each of the selected categories as low selling books may not be representative of the subject category or incorrectly assigned

```python
def waterstones_crawler(url,subject,max_pages):
    '''
    This function crawls through the web page supplied in the function call 
    pulling the required metadata from each book on the page using
    get_book_info() defined below
    '''
    book_data=[]
    page=1
    # Loop through the pages of results for each URL
    while page<=max_pages:   
        # Get all the text from the current page of results
        url1=url+str(page)
        source_code=requests.get(url1,headers)
        plain_text=source_code.text
        plain_text_clean=plain_text.strip()
        soup=BeautifulSoup(plain_text_clean,'html')   
        # Extract the title, synopsis, subject category & ISBN 
        # of each individual book on current page using function get_item_desc
        for link in soup.find_all('a', {'class':'title link-invert dotdotdot'}):
            href='https://www.waterstones.com'+link.get('href')
            book_info=get_book_info(href,subject)
            book_data.append(book_info)  
        page+=1
    return book_data

# Extract the required data for an individual book
def get_book_info(item_url,subject):
    # Get all the text from the current book link
    source_code=requests.get(item_url,headers)
    plain_text=source_code.text
    plain_text_clean=plain_text.strip()
    soup=BeautifulSoup(plain_text_clean)
    # Get title
    title=soup.find('span',{'class':'book-title'}).string
    # Get synopsis
    synopsis=soup.find('div', {'itemprop':'description', 'id':'scope_book_description'}).find_all('p')
    # Get subject category 
    subject=subject
    # Get ISBN
    isbn=soup.find('span',{'itemprop':'isbn'}).string
    return title, synopsis, subject, isbn

# Extract book information for 8 selected subject categoris
spor=waterstones_crawler('https://www.waterstones.com/category/sports-leisure/sortmode/bestselling/page/','sports-leisure',4)
hist=waterstones_crawler('https://www.waterstones.com/category/history/sortmode/bestselling/page/','history',4)
roma=waterstones_crawler('https://www.waterstones.com/category/romantic-fiction/sortmode/bestselling/page/','romantic-fiction',4)
reli=waterstones_crawler('https://www.waterstones.com/category/spirituality-beliefs/sortmode/bestselling/page/','spirituality-beliefs',4)
scie=waterstones_crawler('https://www.waterstones.com/category/science-technology-medicine/sortmode/bestselling/page/','science-technology-medicine',4)  
food=waterstones_crawler('https://www.waterstones.com/category/food-drink/sortmode/bestselling/page/','food-drink',4)
busi=waterstones_crawler('https://www.waterstones.com/category/business-finance-law/sortmode/bestselling/page/','business-finance-law',4)  
ente=waterstones_crawler('https://www.waterstones.com/category/entertainment/sortmode/bestselling/page/','entertainment',4)

# Bring together all data into one dataframe & save as csv
df_spor=pd.DataFrame(spor,columns=['Title','Synopsis','Subject','ISBN'])
df_hist=pd.DataFrame(hist,columns=['Title','Synopsis','Subject','ISBN'])
df_roma=pd.DataFrame(roma,columns=['Title','Synopsis','Subject','ISBN'])
df_reli=pd.DataFrame(reli,columns=['Title','Synopsis','Subject','ISBN'])
df_scie=pd.DataFrame(scie,columns=['Title','Synopsis','Subject','ISBN'])
df_food=pd.DataFrame(food,columns=['Title','Synopsis','Subject','ISBN'])
df_busi=pd.DataFrame(busi,columns=['Title','Synopsis','Subject','ISBN'])
df_ente=pd.DataFrame(ente,columns=['Title','Synopsis','Subject','ISBN'])
list_all = [df_spor, df_hist, df_roma, df_reli, df_scie, df_food, df_busi, df_ente] 
df_all = pd.concat(list_all,ignore_index=True)
df_all.to_csv("C:\Users\jmd05\DSM-020\4. CW1\Data\all.csv")
```

#### a. Clean text fields
Undertake basic cleaning to remove superfluous characters from strings. 

```python
def text_clean(df):
    for i in df.columns:
        # Remove leading & trailing white space & EOF markings
        df[i]=df[i].map(lambda x: x.strip()) 
        # Remove HTML tags
        df[i]=df[i].apply(lambda x: re.sub('<[^<]+?\>',' ',x))
        # Replace punctuation characters with a single white space
        df[i]=df[i].apply(lambda x: x.translate(translator))
        # Replace consecutive white spaces with a single white space
        df[i]=df[i].apply(lambda x: re.sub('\s+',' ',x))
        # Remove leading & trailing white space & EOF markings
        df[i]=df[i].map(lambda x: x.strip()) 
        # Set all features to one case to ensure consistent treatment of characters on-going
        df[i]=df[i].str.lower()                               
        # Remove superfluous book-type descriptor
        df[i] = df[i].map(lambda x: x.replace("paperback","").replace("hardback","")) 
    return df

df1=df.copy()
df1=text_clean(df1)
```

#### c. Treat missing values
Missing synopsis data cannot be inferred so it needs to be populated using alternative sources of data. Extract data from free APIs - Google Books, Wikipedia & OpenLibrary - through the isbnlib package - with ISBN being the match key for the join.

```python
# Check for missing data

dic={}
for i in df1.columns:
    dic[i]=len(df1[df1[i]==''])
print(dic)

# 119/768 (15%) of titles have missing synopsis
# As this is a significant number will need to populate from another data source
```

```python
# Using time.sleep() function here to reduce the possibility of a HTTPError warning (4 minute run time)
def get_synopsis(synopsis, isbn):
    '''
    Check if the synopsis field of a book is missing and if so
    to populate it from one of three open source book libraries
    '''
    if synopsis=='':
            time.sleep(2)
            Service='wiki'
            Wiki=desc(isbn)
            if Wiki!='':
                return Wiki
            else:
                Service='goob'
                Goob=desc(isbn)
                if Goob!='':
                    return Goob
                else:
                    Service='openl'
                    Openl=desc(isbn)
                    if Openl!='':
                        return Openl
                    else:
                        return '' 
    else:
        return synopsis
    
df1['Synopsis']=df1.apply(lambda x: get_synopsis(x['Synopsis'],x['ISBN']), axis=1)

# Clean the data as per (4)(A) above
df2=df1.copy()
df2=text_clean(df2)

# Check population rate
print(len(df2[df2['Synopsis']=='']))
# 19/768 (2%) of titles still have missing synopsis
```

#### d. Split synopsis into individual words & remove redundant words
- Using stop words, remove words which are abundant across all books like 'a' & 'the' and have no unique value for analysis. Also removed the word 'book' which appears in most synopses as an unnecessary reference to itself

```python
# Get standard stop words in English 
stop=stopwords.words('english')

# Split words in each synopsis, remove stop words & remove words containing digits
df2['Synopsis1']=df2['Synopsis'].str.split().apply(lambda x: [item for item in x if item not in stop and item!='book']).apply(lambda x: [item for item in x if item.isalpha()])
```

#### e. Lemmatize words
Here I want to reduce the number of words that comprise the same information. For ease of interpretation we can use a lemmatization technique to stem words to their root word but whereby the root which is an actual word, unlike for other stemming approaches

```python
wordnet_lemmatizer = WordNetLemmatizer()

def lemmatize(x):
    lem_list=[]
    for word in x:
        lem_list.append(wordnet_lemmatizer.lemmatize(word, pos="v"))
    return lem_list

df2['Synopsis2']=df2['Synopsis1'].apply(lambda x: lemmatize(x))
```

#### c. Find how particular a word is to a subject category
- Asides simply comparing the most popular words within each subject category we can use a more sophisticated technique to check if a word is more relevent to one subject category than the other subject categories. We do this by calculating a Term Frequency-Inverse Document Frequency (TF-IDF) score for each word.  


- A word with a high TF-IDF score appears more frequently in a given subject category but not so much in the other subject categories. A more common word like 'the' (if retained in the data) will have a very low TF-IDF score as it tends to appear frequently in all books. 


- The results confirm the differences in the word clouds with the catgeories of 'sports leisure', 'history', 'romantic fiction', 'food drink' & 'entertainment' having words strongly particular to themselves but that the categories of 'spirituality beliefs', 'science technology medicine' & 'business finance law' sharing the same top words of 'life', 'world' and 'time'.

```python
# Convert dictionary into a dataframe
words_cat = pd.DataFrame(list(dic0.items()),columns=['Subject','All words'])
words_cat['All words']=words_cat['All words'].astype("string")

# Calculate the TF-IDF scores for each word in each subject category

def tfidf(cats,i,j):
    
    # Calculate Term Frequency (TF) values
    # Instantiate CountVectorizer() - single character words are ignored by tokeniser
    cv=CountVectorizer() 
    # Generate word frequency sparsity matrix
    word_count_vector=cv.fit_transform(cats)

    # Calculate the Inverse Document Frequency (IDF) weight values
    # Instantiate the IDF transformer
    tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True) 
    # Fit to the sparsity matrix
    tfidf_transformer.fit(word_count_vector)

    # Calculate the TF x IDF (TFIDF) weight values
    tf_idf_vector=tfidf_transformer.transform(word_count_vector)

    # Get TF-IDF scores for ith subject category
    ith_document_vector=tf_idf_vector[i] 
    df = pd.DataFrame(ith_document_vector.T.todense(), index=cv.get_feature_names(), columns=[str(j)+"-TFIDF"]) 
    df1=df.sort_values(by=[str(j)+"-TFIDF"],ascending=False)
    df1[str(j)+"-word"] = df1.index
    df1.reset_index(drop=True, inplace=True)
    df2=df1[[str(j)+"-word",str(j)+"-TFIDF"]]
    
    return(df2)

spor=tfidf(words_cat.iloc[:,1],0,'Spor').iloc[:10]
hist=tfidf(words_cat.iloc[:,1],1,'Hist').iloc[:10]
roma=tfidf(words_cat.iloc[:,1],2,'Roma').iloc[:10]
spir=tfidf(words_cat.iloc[:,1],3,'Spir').iloc[:10]
scie=tfidf(words_cat.iloc[:,1],4,'Scie').iloc[:10]
food=tfidf(words_cat.iloc[:,1],5,'Food').iloc[:10]
busi=tfidf(words_cat.iloc[:,1],6,'Busi').iloc[:10]
ente=tfidf(words_cat.iloc[:,1],7,'Ente').iloc[:10]

# Bring together
all_dfs = [spor, hist, roma, spir, scie, food, busi, ente]
tfidf_all = reduce(lambda  left,right: pd.merge(left, right, left_index=True, right_index=True, how='outer'), all_dfs)
tfidf_all
```

#### d. Find how particular a phrase is to a subject category
- As phrases contain more contextual information than single words, the TF_IDF methodology can also be applied to two-word phrases (bigrams) to look for differences across the subject categories. 


- As expected the results show clearer differentiation across the subject categories compared to the above analysis using single words. 


- However there are generic phrases like 'Non fiction' & 'bestselling author' appearing in a couple of subject categories, most notably in the 'science, technology, medicine' category, which while differentiating they are not the most meaningful phrases expected for their particular subject categories and so some further data cleaning might be required.

```python
def get_most_frequent_phrases(df,j):   
    vectorizer = TfidfVectorizer(ngram_range = (2,2))
    corpus=vectorizer.fit_transform(df.iloc[:,5])
    scores = (corpus.toarray())
    features = (vectorizer.get_feature_names())
    sums = corpus.sum(axis = 0)
    data = []
    for col, term in enumerate(features):
        data.append((term, sums[0,col]))
    ranking = pd.DataFrame(data, columns=['term','rank'])
    words = (ranking.sort_values('rank',ascending=False)).reset_index(drop=True).iloc[:10]
    words.rename(columns = {'term':str(j)+"-ngram", 'rank':str(j)+"-TFIDF"}, inplace = True)
    return words

spor=get_most_frequent_phrases(df3[df3['Subject']=='sports leisure'],'spor')
hist=get_most_frequent_phrases(df3[df3['Subject']=='history'],'hist')
roma=get_most_frequent_phrases(df3[df3['Subject']=='romantic fiction'],'roma')
spir=get_most_frequent_phrases(df3[df3['Subject']=='spirituality beliefs'],'spir')
scie=get_most_frequent_phrases(df3[df3['Subject']=='science technology medicine'],'scie')
food=get_most_frequent_phrases(df3[df3['Subject']=='food drink'],'food')
busi=get_most_frequent_phrases(df3[df3['Subject']=='business finance law'],'busi')
ente=get_most_frequent_phrases(df3[df3['Subject']=='entertainment'],'ente')

# Bring together
all_dfs_ngrams = [spor, hist, roma, spir, scie, food, busi, ente]
tfidf_all_ngrams = reduce(lambda  left,right: pd.merge(left, right, left_index=True, right_index=True, how='outer'), all_dfs_ngrams)
tfidf_all_ngrams
```

#### e. Measure the overall similarities of subject categories
- There are around 11.5K unique words across all the subject categories and so an 11.5K dimensional array (vector) of word frequencies for each subject category can be created. Euclidean distances between each pair of vectors can be calculated to find out how far apart each pair is in this space. This distance measure can act as a proxy for how similar two subject catgeories are in terms of vocabulary so rather than comparing the top single words or phrases across subject categories this technique enables us to compare all words in a category at once for an overall measure of similarity. A distance of 0 between two vectors implies that both have exactly the same words and their frequencies. 


- The distance results in the heat map below show that the darker the colour of the square the smaller the distance between two subject categories. As expected there are are significant differences in distances between obviously diverse subjects like 'sport leisure' and 'romantic fiction' (distance=307) and 'romantic fiction' and 'history' (distance=302) meaning their vocabularies are significantly different. 'food drink' has the largest distances from other subject categories so its vocabulary has the least in common with the others. Notably the pairwise combinations of 'spirituality beliefs', 'science technology medicine' & 'business finance law' show the most similar vocabularies out of the 8 subject categories concurring with findings above.

```python
# Get data
spor=' '.join(dic0['sports leisure'])
hist=' '.join(dic0['history'])
roma=' '.join(dic0['romantic fiction'])
spir=' '.join(dic0['spirituality beliefs'])
scie=' '.join(dic0['science technology medicine'])
food=' '.join(dic0['food drink'])
busi=' '.join(dic0['business finance law'])
ente=' '.join(dic0['entertainment'])
all=[spor,hist,roma,spir,scie,food,busi,ente]

# Create distance matrix
with warnings.catch_warnings():
    # Ignore warning for this code
    warnings.filterwarnings("ignore")
    cv = CountVectorizer(all)
    words=cv.fit_transform(all)

words=words.toarray()
words.shape
#(8, 11558)

dist_matrix=distance_matrix(words,words)
dist_matrix_df=pd.DataFrame(dist_matrix, index=subjects,columns=subjects).astype(int)
#dist_matrix_df
```
